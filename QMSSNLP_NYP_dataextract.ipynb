```python
year_list = list(range(1851,2022))
month_list = list(range(1,13))
key = 'FBp0wuqY2U1vvTq4GY21pGJuUhB2KKVO'
url = 'https://api.nytimes.com/svc/archive/v1/2019/1.json?api-key='+key
r = requests.get(url)
json_data = r.json()
type_list = []
data_list = []
num_list = []
for year in year_list:
    for month in month_list:
        root_url = 'https://api.nytimes.com/svc/archive/v1/' + str(year)+'/'+ str(month)+'.json?api-key='
        url = root_url+key
        time.sleep(8)
        r = requests.get(url)
        json_data = r.json()
        num_docs = pyjq.all('.response .docs | length',json_data)[0]
        data = pyjq.all('.response',json_data)
        data_list.append(data)
        num_list.append(num_docs)
        for i in range(num_docs):
            type_list.append(data[0]['docs'][i]['type_of_material'])
        print('finished_year: ' + str(year)+' finished_month: '+ str(month) )
        
```
